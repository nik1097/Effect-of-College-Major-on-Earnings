### Dataset 

We'll be working with a dataset on the job outcomes of students who graduated from college between 2010 and 2012. The original data on job outcomes was released by [American Community Survey](https://www.census.gov/programs-surveys/acs/), which conducts surveys and aggregates the data. 
The cleaned data can be found [here](https://github.com/fivethirtyeight/data/tree/master/college-majors)

```python
recent_grads = pd.read_csv('recent-grads.csv')
```

### Exploring the Data

```python
print(recent_grads.iloc[0])
print(recent_grads.head())
print(recent_grads.tail())
recent_grads = recent_grads.dropna()
```

### Scatter Plots

```python
ax1 = recent_grads.plot(x='Sample_size', y='Median',kind='scatter', title='Sample_size and Median')
```

Lesser the number of Sampled Full Time Employees(Sample_size) graduated from a major, more is the Median Salary. We will be considering Sample_size and not Full_time as the latter was used to calculate Earnings in the dataset.

```python
ax2 = recent_grads.plot(x='Sample_size', y='Unemployment_rate',kind='scatter', title='Sample_size and Unemployment_rate')
```

Higher the number of Sampled Full Time Employees(Sample_size)graduated a major, lower is the Unemployment Rate. We will be considering Sample_size and not Full_time as the latter was used to calculate Earnings in the dataset.

```python
ax = recent_grads.plot(x='Full_time', y='Median',kind='scatter', title='Full_Time and Median')
```

The trend for Sample_size and Median Salary is almost the same as Full_Time and Median Salary, lesser the number of Full Time employees graduated from a major, more is the Median Salary. This also verifies the Sampling was accurate. 

```python
ax = recent_grads.plot(x='ShareWomen', y='Unemployment_rate',kind='scatter', title='ShareWomen and Unemployment_rate')
```

The Women as Share of Total and Unemployment Rate don't exhibit a trend as it is shown in the graph, it is difficult to obtain a relation from this graph.

```python
ax = recent_grads.plot(x='Men', y='Median',kind='scatter', title='Men and Median')
```

Lower the number of Male Graduates, more is the Median Salary, this asserts the fact that lesser the number of graduates from a major, more higher is the Median salary.

```python
ax = recent_grads.plot(x='Women', y='Median',kind='scatter', title='Women and Median')
```

Lower the number of Female Graduates, more is the Median Salary.This asserts the fact that lesser the number of graduates from a major, more higher is the Median salry.

### Bar Graphs

Comparing percentages of women (ShareWomen) from the first ten rows and last ten rows of the recent_grads dataframe.

```python
recent_grads[:10].plot.bar(x='Major', y='ShareWomen', legend=False)
recent_grads[-10:].plot.bar(x='Major', y='ShareWomen', legend=False)
```

The last 10 rows have a much higher percentage of women in their majors than the first 10 rows.

Comparing the unemployment rate (Unemployment_rate) from the first ten rows and last ten rows of the recent_grads dataframe.

```python
recent_grads[:10].plot.bar(x='Major',y='Unemployment_rate')
recent_grads[-10:].plot.bar(x='Major',y='Unemployment_rate')
```

There is very little difference between the Unemployment Rates in the first 10 rows and the last 10 rows.

### Grouped Bar Graphs

```python
recent_grads[['Major','Men','Women']][:10].plot.bar(x='Major')
recent_grads[['Major','Men','Women']][-10:].plot.bar(x='Major')
```

It can be seen that the last 10 majors have a higher number of women and the first 10 majors have a higher number of men.
